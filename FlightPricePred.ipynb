### Feature Engineer
- Try to change the object type into numerical form as much as possible
- Then change the categorical into numerical

```python
#Missing value
final_df.isnull().sum()
#price has NaN value in the Test_set appended at the end of Train Set
```

```python
final_df[final_df['Route'].isnull()]
```

```python
final_df.dropna(subset=['Route'],axis=0,inplace=True)
```

```python
final_df.isnull().sum()
```

```python
final_df['Date_of_Journey'].str.split('/')[0]
```

```python
final_df['Date_of_Journey'].str.split('/').str[0]#date
```

```python
final_df['Date_of_Journey'].str.split('/').str[1]#Month
```

In the case of forecasting using date this tecnique must not used

```python
#Month, Date, Year
final_df['Date']=final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month']=final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year']=final_df['Date_of_Journey'].str.split('/').str[2]
#or
#df['Date_of_Journey'].apply(lambda x:x.split("/")[0])
```

```python
final_df['Date']=final_df['Date'].astype(int)
final_df['Month']=final_df['Month'].astype(int)
final_df['Year']=final_df['Year'].astype(int)
```

```python
final_df.info()
```

```python
#droping date_of_journey; no need of keeping it
final_df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
final_df.head(1)
```

```python
#next is Arrival_time
final_df['Arrival_Time']=final_df['Arrival_Time'].apply(lambda x : x.split(' ')[0])
```

```python
final_df['Arrival_hour']=final_df['Arrival_Time'].apply(lambda x : x.split(':')[0])
final_df['Arrival_min']=final_df['Arrival_Time'].apply(lambda x : x.split(':')[1])
```

```python
final_df['Arrival_hour']=final_df['Arrival_hour'].astype(int)
final_df['Arrival_min']=final_df['Arrival_min'].astype(int)
final_df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
final_df.head(1)
```

```python
final_df.info()
```

```python
final_df.head()
```

```python
# next is Dep_time
final_df['Dep_hour']=final_df['Dep_Time'].apply(lambda x : x.split(':')[0])
final_df['Dep_min']=final_df['Dep_Time'].apply(lambda x : x.split(':')[1])
```

```python
final_df['Dep_hour']=final_df['Dep_hour'].astype(int)
final_df['Dep_min']=final_df['Dep_min'].astype(int)
final_df.drop('Dep_Time',axis=1,inplace=True)
```
