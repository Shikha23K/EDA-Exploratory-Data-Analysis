```python
df_country=pd.read_excel('D:\corizo\data\zomato\Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
#combine xomato csv feature country_code with the data frame country-code.xlsx
final_df=pd.merge(df,df_country,on='Country Code', how='left')#can use concat also
final_df.head(3)
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

```python
country_name=final_df.Country.value_counts().index
```

```python
country_value=final_df.Country.value_counts().values
```

```python
#which country has done maximum trnsaction

# Pie chart
plt.pie(country_value,labels=country_name)
```

```python
#or top 3 country in transaction
plt.pie(country_value[:3],labels=country_name[:3])
```

```python
plt.pie(country_value[:3],labels=country_name[:3],autopct="%1.2f%%")
```

Observation: Zomato maximum records or transaction are from India after that USA and then UK

```python
#Pandas groupby is used for grouping the data according to the categories and apply a function to the categories.
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
#size() returns count of each record group
#ratings.reset_index()#it will return a dataframe with reset index and new column with name'0'

```

```python
ratings.reset_index().rename(columns={0:'Rating Count'})#rename the new column '0'
```

Observation :
- When rating is between 4.5 to 4.9 --->Excellent
- when ratings are between 4.0 to 4.4--->very good
- when rating is between 3.5 to 3.9 --->good
- when rating is between 3.0 to 3.4 --->average
- when rating is between 2.5 to 2.9 --->average
- when rating is 2.0 t0 2.4---->poor
and so on.

plot the ratings.

```python
ratings.columns
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating count',data=ratings)
```

```python
# Mapping the rating color given in dataset to the bars
sns.barplot(x='Aggregate rating',y='Rating count',hue='Rating color',data= ratings)
```

Here the colored bar is different from the rating color records so we will give the same color using palette parameter

```python
sns.barplot(x='Aggregate rating',y='Rating count',hue='Rating color',data= ratings,palette=('grey','red','orange','yellow','green','green'))
```

Observation: 
- Not rated count is very high
- Maximum no of ratings are between 2.5 to 4.4

```python
#Count plot
sns.countplot(x="Rating color",data=ratings,palette=('blue','red','orange','yellow','green','green'))
```

count represents how frequently the color is given for ratings. It is not rating count
